```python
trial = data[['InvoiceNo','CustomerID','InvoiceDate','StockCode','Quantity','Warranty','UnitPrice']]
trial = trial.set_index(['CustomerID', 'InvoiceNo','StockCode'])
trial.head()
```

```python
trial['Price'] = trial['Quantity']*trial['UnitPrice']
trial.head(20)
trial.to_csv('OrderInvoice.csv')
```

```python
import re
from datetime import datetime
dt = datetime.now()
new_list = []
for i in range(0,len(trial)):
    new_list.append(dt.year - int(re.split('/| |:',trial['InvoiceDate'].iloc[i])[2]))
```

```python
trial['Age'] = new_list
trial.head()
```

```python
def recommend_warranty(cus_id,pred):
    new_df = trial.loc[cus_id]
    new_df['WarrantyStatus'] = new_df['Age']-new_df['Warranty']
    #new_df['out_war'] = (lambda x: (x>0), 1, 0)
    #compare = lambda x: 1 if x>0 else 0
    new_df['WarrantyStatus'] = new_df.apply(lambda row: 1 if (row[6]<0) else 0, axis = 1)
    out_war = new_df['WarrantyStatus'] == 0
    final_df = new_df[out_war].sort_values(['Quantity','Price'],ascending=[False,True])
    list_prod = []
    for i in range(0,pred):
        list_prod.append(final_df.index[i][0])
    return list_prod
recommend_warranty(17850.0,10)
```
